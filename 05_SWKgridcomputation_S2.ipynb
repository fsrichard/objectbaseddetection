```python
from datacube.utils import geometry
geom = geometry.Geometry(geom=districts_gdf.iloc[0].geometry,
                         crs=districts_gdf.crs)
```

```python
dc = datacube.Datacube(app='Using_load_ard')
```

```python
%%time

from deafrica_tools.datahandling import load_ard

# Create a reusable query
query = {
    'time': ('2019-01', '2019-12'),
    'measurements': ['B02', 'B03', 'B04', 'nir'],
    'output_crs': 'epsg:6933',
    'resolution': (-10, 10),
    'group_by': 'solar_day',
    'dask_chunks': {"x": 2000, "y": 2000}
}

# Load available data from S2
s2 = load_ard(dc=dc, geopolygon=geom,
              products=['s2_l2a'],
              **query)

# Print output data
s2
```

```python
dcf = s2.to_array(dim="variable")

dcf 
```

```python
median = dcf.to_dataset(dim="variable") #important to explicitly define dim

median
```

```python
# Compute a single median from all data
ds_median = median.median('time')

ds_median
```

```python
data_meanarray = ds_median.to_array()
data_meanarray
```

```python
data = data_meanarray.to_dataset(dim='variable')  
data
```

```python
#vegetation indices 

# Calculate the NDVI
band_diff = data.nir - data.B04
band_sum = data.nir + data.B04
ndvi = band_diff / band_sum

#layer ndvi to the data
data['ndvi']=ndvi
data
```

```python
#EVI layer
#sample from earth engine 

"""var evi = comp.expression(
    '2.5 * ((NIR - RED) / (NIR + 6 * RED - 7.5 * BLUE + 1))', {
      'NIR': comp.select('B5'),
      'RED': comp.select('B4'),
      'BLUE': comp.select('B2')
}); """
evi = (2.5 * ((data.nir - data.B04) / (data.nir + 6 * data.B04 - 7.5 * data.B02 +1)))

data['evi']=evi
data
```

```python
#In Landsat 8, SAVI = ((Band 5 – Band 4) / (Band 5 + Band 4 + 0.5)) * (1.5)
#from GEE, L8
"""var savi = comp.expression(
    '((B5 - B4) / (B5 + B4 + 0.5)) * (1.5)',{
      'B5': comp.select('B5'),
      'B4': comp.select('B4')
      """

savi = (((data.nir - data.B04)/(data.nir + data.B04 + 0.5)) * 1.5) 
data['savi']= savi 
data
```

```python
# GCI 
#from GEE

"""var gci = comp.expression(
    'NIR / (B3 - 1)',{
      'NIR': comp.select('B5'),
      'B3': comp.select('B3')
});"""

gci = ((data.nir / data.B03) - 1) 

data['gci']= gci
data
```

```python
# Rename some of the bands to match the bands during the model development. 
data['B2'] = data['B02']  
data = data.drop(['B02']) 
```

```python
data['B3'] = data['B03']  
data = data.drop(['B03']) 
```

```python
data['B4'] = data['B04']  
data = data.drop(['B04']) 
```

```python
data['B8'] = data['nir']  
data = data.drop(['nir']) 

data
```

```python
#Rearrange to match the model's features.
data = data[['B2', 'B3', 'B4', 'B8', 'evi', 'gci', 'ndvi', 'savi']]  

data
```

```python
# Import the trained model. 
import joblib
model = joblib.load("model/HGBtuned_S2OP.pkl")

print(model)
```

```python
# Choose file containing test areas and load
districts_file = "swk_bahagian/Division_boundary.shp"
districts_gdf = gpd.read_file(districts_file)

# Set results path
results_path = "figures/"

districts_gdf
```

```python
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
# "UserWarning: X does not have valid feature names, but HistGradientBoostingClassifier was fitted with feature names"

# filter the shape files per district based on its ID 
area_of_interest_gdf = districts_gdf.iloc[[0]]
district_column = "KM2"

for index, district in area_of_interest_gdf.iterrows():
    
    # Set up geometry
    district_name = str(int(district[district_column]))
    print(f"Processing {district_name}")
    
    # Check if district has already been processed. If so, skip
    output_filename = f"{results_path}/district_{district_name}_oilpalmprediction.tif"
    if os.path.exists(output_filename):
        print("Completed; Skipping")
        continue

    # set up query based on district polygon
    geom = geometry.Geometry(geom=district.geometry, crs=area_of_interest_gdf.crs)
    query.update({"geopolygon": geom})

    #predict using the imported model
    predicted = predict_xr(model,
                           data.unify_chunks(),
                           #proba=True,
                           persist=True,
                           clean=True,
                           return_input=False
                          ).astype(float).persist()
    
    # Write to cog
    prediction_file = f"{results_path}/district_{district_name}_oilpalm_prediction.tif"
    print(f"    Writing predictions to {prediction_file}")
    write_cog(
        predicted.Predictions.compute(),
        fname=prediction_file,
        overwrite=True
    )
    
    del predicted
```
